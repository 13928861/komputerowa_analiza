```python
#zad2 podpunkt 4
import numpy as np
import matplotlib.pyplot as plt

mu = 0
sigma_2= 1

MC = 200  
długość_próbki= [50, 100, 1000]  

def MM(Y):
    return (np.mean(Y) * np.sqrt(np.pi)) / np.sqrt(2)

def MNW(Y):
    return np.sqrt(np.mean(Y**2))

mm2 = {n: [] for n in długość_próbki}
mnw2 = {n: [] for n in długość_próbki}


for n in długość_próbki:
    for _ in range(MC):
        X = np.random.normal(mu, sigma_2, n)
        Y = np.abs(X - mu)
    
        mm3 = MM(Y)
        mnw3  = MNW(Y)
        
        mm2[n].append(mm3)
        mnw2[n].append(mnw3)


fig, ax = plt.subplots(1, 2, figsize=(10, 5))

ax[0].boxplot([mm2[n] for n in długość_próbki], labels=długość_próbki)
ax[0].axhline(sigma_2, color='brown', linestyle='--', label=f'σ = {sigma_2}')
ax[0].set_title('Metoda momentów')
ax[0].set_xlabel('wielkość próbki(n)')
ax[0].set_ylabel('wartość estymatora')
ax[0].legend()

ax[1].boxplot([mnw2[n] for n in długość_próbki], labels=długość_próbki)
ax[1].axhline(sigma_2, color='brown', linestyle='--', label=f'σ = {sigma_2}')
ax[1].set_title('Metoda największej wiarygodności')
ax[1].set_xlabel('wielkość próbki')
ax[1].set_ylabel('wartość estymatora')
ax[1].legend()

plt.tight_layout()
plt.show()
```
